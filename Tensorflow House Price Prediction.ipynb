```python
import tensorflow as tf
```

```python
import pandas as pd
```

```python
housing =pd.read_csv('house_pricing.csv')
```

```python
housing.head()
```

```python
housing.describe().transpose()
```

```python
xdata=housing.drop(['medianHouseValue'],axis=1)
```

```python
yval=housing['medianHouseValue']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(xdata, yval, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train=pd.DataFrame(data=scaler.transform(X_train),columns=X_train.columns,index=X_train.index)
```

```python
X_test=pd.DataFrame(data=scaler.transform(X_test),columns=X_test.columns,index=X_test.index)
```

```python
housing.columns
```

```python
X_train.head()
```

```python
age=tf.feature_column.numeric_column('housingMedianAge')
rooms=tf.feature_column.numeric_column('totalRooms')
bedrooms=tf.feature_column.numeric_column('totalBedrooms')
pop=tf.feature_column.numeric_column('population')
households=tf.feature_column.numeric_column('households')
income=tf.feature_column.numeric_column('medianIncome')
```

```python
featcols=[age,rooms,bedrooms,pop,households,income]
```

```python
inputfunc=tf.estimator.inputs.pandas_input_fn(x=X_train,y=y_train,batch_size=10,num_epochs=1000,shuffle=True)
```

```python
model=tf.estimator.DNNRegressor(hidden_units=[6,6,6],feature_columns=featcols)
```

```python
model.train(input_fn=inputfunc,steps=2500)
```

```python
predictinputfunc=tf.estimator.inputs.pandas_input_fn(x=X_test,batch_size=10,num_epochs=1,shuffle=False)
```

```python
predgen=model.predict(predictinputfunc)
```

```python
predictions=list(predgen)
```

```python
predictions
```

```python
finalpred=[]
for pred in predictions:
    finalpred.append(pred['predictions'])
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,finalpred)**0.5
```
